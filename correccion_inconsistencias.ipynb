```python
import pandas as pd
import seaborn as sns

# archivo .xlsx
df_productos = pd.read_excel("Assets/productos.xlsx")


# 1. Definir las palabras clave para cada categoría
palabras_clave_alimentos = ['yogur', 'leche', 'pan', 'fruta', 'carne', 'verdura', 'queso', 'cereal', 'yerba', 'jugo', 'café', 'té', 'galleta', 'snack', 'agua', 'refresco', 'chocolate', 'dulce', 'miel', 'harina', 'arroz', 'pasta', 'aceite', 'vinagre', 'especia', 'salsa', 'alfajor', 'lenteja', 'garbanzo', 'frijol', 'nuez', 'semilla', 'fruto seco', 'embutido', 'jamón', 'salchicha', 'pollo', 'pescado', 'pepsi', 'coca cola', 'fanta', 'sprite', 'gaseosa', 'bebida energética', 'bebida isotónica', 'yogur bebible', 'yogur griego', 'queso fresco', 'queso rallado', 'pan integral', 'pan de molde', 'pan baguette', 'fruta fresca', 'verdura fresca', 'carne vacuna', 'carne de cerdo', 'carne de pollo', 'cereal integral', 'cereal azucarado', 'jugo natural', 'jugo en polvo','papas fritas', 'snack salado', 'snack dulce','turrón', 'chicle', 'golosina', 'stevia','bebida cola', 'bebida saborizada','azucar','avena','vino','cerveza','whisky','ron','vodka','tequila']
palabras_clave_limpieza = ['jabón', 'detergente', 'limpiador', 'desinfectante', 'cloro', 'escoba', 'trapeador', 'esponja', 'guante', 'spray']

# 2. Asignación Masiva: Asignar a 'Alimentos'
# Usa str.contains para buscar cualquiera de las palabras clave en el nombre del producto (ignorando mayúsculas)
condicion_alimentos = df_productos['nombre_producto'].str.lower().str.contains('|'.join(palabras_clave_alimentos), na=False)

df_productos.loc[condicion_alimentos, 'categoria'] = 'Alimentos'


# 3. Asignación Masiva: Asignar a 'Limpieza'
# Usamos el mismo método para Limpieza
condicion_limpieza = df_productos['nombre_producto'].str.lower().str.contains('|'.join(palabras_clave_limpieza), na=False)

df_productos.loc[condicion_limpieza, 'categoria'] = 'Limpieza'


# 4. Verificar la categorización de una muestra aleatoria
print("\nVerificación de Categorías Corregidas:")
print(df_productos.sample(20))  # Muestra 20 filas aleatorias para verificar las categorías asignadas
# Guardar el DataFrame con las categorías correctas
#df_productos.to_excel("productos_corregidos.xlsx", index=False)
```
