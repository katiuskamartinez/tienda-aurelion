```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# archivo .xlsx
df = pd.read_excel('Assets/detalle_ventas.xlsx')

columna_analizar = 'importe' # O la columna que se elija (cantidad, precio_unitario, etc.)

# Visualización de la Distribución
# a) Calcular Estadísticos Clave
media = df[columna_analizar].mean()
mediana = df[columna_analizar].median()
moda = df[columna_analizar].mode()
asimetria = df[columna_analizar].skew()

print(f"Media de '{columna_analizar}': {media:.2f}")
print(f"Mediana de '{columna_analizar}': {mediana:.2f}")
print(f"Asimetría (Skewness): {asimetria:.2f}")

if abs(asimetria) < 0.5:
    print("-> Distribución probablemente Simétrica (Normal o Uniforme).")
elif asimetria > 0.5:
    print("-> Distribución con Asimetría Positiva (Sesgada a la Derecha, ej. Ingresos).")
else: # asimetria < -0.5
    print("-> Distribución con Asimetría Negativa (Sesgada a la Izquierda).")


    # Crear el Histograma (Visualización de Frecuencias)
plt.figure(figsize=(10, 6))
sns.histplot(df[columna_analizar], kde=True, bins=30)
plt.title(f'Histograma y Densidad de {columna_analizar}')
plt.axvline(media, color='red', linestyle='dashed', linewidth=1, label='Media')
plt.axvline(mediana, color='green', linestyle='dashed', linewidth=1, label='Mediana')
plt.legend()
plt.show()
```
