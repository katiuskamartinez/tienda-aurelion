# Punto 5 — Interpretación de resultados

A continuación se resumen e interpretan los hallazgos principales derivados de los gráficos generados (distribución de clientes por ciudad, relación cantidad vs. importe y distribución de precio unitario por categoría).

## 1) Distribución de clientes por ciudad
- Observación: se identifican ciudades con concentración alta de clientes y otras con muy pocos clientes.
- Interpretación: la demanda está focalizada en unas pocas ciudades; puede indicar foco de operaciones o desigualdad en la cobertura.
- Acción recomendada: priorizar acciones comerciales y logística en las ciudades con mayor volumen; investigar causas de baja representación en otras ciudades (acceso, oferta, datos incompletos).

## 2) Relación cantidad de productos vendidos vs. importe total
- Observación: existe una tendencia positiva (a mayor cantidad, mayor importe) pero con dispersión y outliers.
- Interpretación: la relación esperada ocurre en general; los outliers donde importe es alto con baja cantidad pueden indicar productos de alto precio o errores/errores de ingreso.
- Acción recomendada: calcular la correlación (Pearson) y R² para cuantificar la relación; revisar ventas atípicas para detectar precios erróneos o transacciones especiales.

## 3) Distribución del precio unitario por categoría
- Observación: diferencias en medianas y rangos intercuartílicos entre categorías; presencia de valores extremos (outliers).
- Interpretación: algunas categorías tienen mayor variabilidad de precio (posible mezcla de productos económicos y premium), otras están más homogéneas.
- Acción recomendada: segmentar precios por subcategoría, revisar outliers (posibles errores o SKUs premium), y calibrar estrategia de precios según categoría.

## Limitaciones
- Calidad y completitud de datos: valores faltantes, formatos de fecha o errores en ingreso pueden sesgar las gráficas.
- Contexto: faltan variables clave (margen, costo, promociones, canal de venta) para conclusiones comerciales firmes.

## Próximos pasos sugeridos
1. Calcular métricas cuantitativas: correlaciones, R², estadísticas por grupo (mediana, IQR).
2. Investigar outliers y limpiar datos (fechas inválidas, importes negativos, cantidades inconsistentes).
3. Enriquecer datos con variables adicionales (canal, costo, fecha de campaña) y repetir análisis.
4. Crear un dashboard interactivo para monitorizar las métricas clave por ciudad y categoría.
