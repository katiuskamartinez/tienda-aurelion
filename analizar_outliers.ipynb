```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# archivo .xlsx
df = pd.read_excel('Assets/detalle_ventas.xlsx')
sns.boxplot(y=df['importe'])
plt.title('Box Plot para detección de Outliers')
plt.show()
```

```python
Q1 = df['precio_unitario'].quantile(0.25)
Q3 = df['importe'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtrar y mostrar los outliers:
outliers = df[
    (df['precio_unitario'] < limite_inferior) | 
    (df['importe'] > limite_superior)
]

print("\nValores Atípicos Encontrados (Outliers):\n", outliers)
```
